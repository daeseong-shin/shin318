# <주식트레이너 귤맨 강의 내용>
## > MTS, HTS 비교
## > 정치관련주, 바이오주, 동전주 거래 하지 말 것
## > 적정 시가총액 확인 방법
### -> 최근 1년 순수익의 10배 = 시가총액 유사 수준 확인(대략적인 방법)
### ---------------------------------------------------------------------------------------------------------
# ★ 단타 매매법(~2일內) / ~3% 內 수익 실현 목표
### 1) 네이버 금융 -> 코스닥 지수 확인
### 2) 거래대금 1,000억 이상(=대장주 확인)
### 3) 박스권 돌파 확인(기존 박스권 돌파보다 많은 물량 必, 박스권 기간 3개월 이상 추천)
### 4) 다음날 2~3% 이익 실현 후 매도(단, 더 상승 할 것으로 예상되면 절반 매도)
### ---------------------------------------------------------------------------------------------------------
# ★ 잠입매매법(=Swing) /  10% 이상 수익 실현 목표
### 1) 기존 Box권 크게 돌파
### 2) 기존 Peek 보다 거래량 많아야 함
### 3) 다음날 기존 Peek 보다 높으면서, 매수 물량의 1/10 수준으로 매도 일 경우
### 4) 신규 Peek의 시가, 종가를 4분할 하여 분할 매수 후 상승 대기
### ---------------------------------------------------------------------------------------------------------
