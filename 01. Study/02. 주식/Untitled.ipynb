```python
# Self Study 내용
## 1. MTS, HTS 차이
### * MTS : Mobile Trading System
### * HTS : Home Trading System
## ------------------------------------------------------------------------------------
## 2. 보통주, 우선주
### * 보통주 : 보통의 주식, 일상에서 부르는 주식
###  -> 주주평등의 원칙에 따라 평등하게 배당받는 주식의 형태
### * 우선주 : 보통주보다 우선권을 갖는 주식
###  -> 주주총회에서 의결권 행사할 수 없음, 보통주보다 배당금 높음
## ------------------------------------------------------------------------------------
## 3. 우량주, 테마주
### * 우량주 : 업적과 경영 내용, 배당률이 높은회사 주식
###  -> 성장가치가 높고, 안전한 주식
### * 테마주 : 증권시장에서 큰 영향을 주는 정치, 사회적 잇슈와 관련된 주식
###  eX) 정치 테마주, 계절 테마주, 대북 테마주, 코로나 테마주
## ------------------------------------------------------------------------------------
## 4. 어닝시즌(Earning season)
###  -> 회사가 분기 또는 반기 실적을 발표하는 시기
###  -> 우리나라의 경우 어닝시즌은 보통 12월 결산법인들의 분기실적이 발표되는 시기
## ------------------------------------------------------------------------------------
## 5. 어닝쇼크(Earning shock) ↔ 어닝서프라이즈(Earning surprise)
###  -> 기업이 시장에서 예상했던 것 보다 저조한 실적을 발표하여 주가에 영향을 미치는 현상
## ------------------------------------------------------------------------------------
## 6. 어닝서프라이즈(Earning surprise) ↔ 어닝쇼크(Earning shock) 
###  -> 어닝쇼크의 반대. 예상보다 높은실적이 발표되 놀라는 현상
## ------------------------------------------------------------------------------------
## 7. 턴어라운드(Turn around) 
###  -> 상황이 좋지 않던 기업이 급작스레 경영능력이 향상 or 매출이 급등하여 이익으로 변동
## ------------------------------------------------------------------------------------
## 8. 모멘텀(Momentum) 
###  -> 주가 상승세 혹은 하락세의 급격한 변화를 나타내는 지표
###  ex) 모멘텀이 지속된다 = 상승추세가 유지
###  ex) 모멘텀이 부족하다 = 상승의 원동력이 부족
## ------------------------------------------------------------------------------------
## 9. 유상증자, 무상증자
### 증자 : 주식 수를 늘리는 것
### -> 일반적으로 주식의 가치는 내려감
## 1) 유상증자
### a. 주주배정 유상증자
###  : 새롭게 발행되는 신주를 기존에 주식을 보유하고 있던 주주들에게 배정하는 방식
###    주로 튼튼하고 주주들에게 신뢰가 높은 우량기업에서 주로 사용 하는 방식
###    -> 약한 악재
### b. 주주우선 공모 유상증자
### : 기존 주주와 우리사주조합에게 새롭게 발행되는 주식을 청약받아 배정하는 방식
###   이후 남은 물량을 일반 투자자들에게 공개적으로 모집하는 방식
###   -> 조금 심한 악재
### c. 일반공모 유상증자
### : 기존 주주에게 신주인수권을 주지 않는 유상증자
###   불특정 다수에게 공개적으로 모집하는 방식
###   -> 가장 큰 악재
### d. 3자배정 유상증자
### : 기존 주주에게 신주인수권을 주지 않음. 단 이미 정해져 있는 제 3자에게 주식을 주는 방식
###   기업의 지분을 취득하거나, 금액이 큰 경우엔 최대 주주가 바뀌는 일도 발생
###   -> 유산증자 중에서 유일하게 시장에서 좋은 평가를 받는 방법
## 2) 무상증자
### : 이익잉여금 만큼 주식을 발행하고, 이를 기존 주주들이 가지고 있는 지분에 비례해 주식을 나눠주는 방식
### : 자기자본의 총액은 변함이 없지만, 무상증자를 하면 기존 주주들은 주식을 공짜로 받게 됨
###   -> 호재
## ------------------------------------------------------------------------------------
## 10. 예수금, 증거금, 미수금, 대융금
## [예수금] = 현금
### = 1. 주식을 살 수 있는 돈, 2. 현금화 후 계좌에 들어 오는 돈
## [증거금] = 약정대금의 일정비율에 해당하는 금액을 미리 예탁해야하는 보증금
### ex) 증거금률 50% 일 경우 실제 매수 금액의 50%만으로도 매수 가능. 단 앞으로 2일 안에 50% 예수금에서 빠짐
###     증거금률 100% = 내 현금으로만 매매하는 것
## [미수금] = 결제일에 결제대금을 납부하지 못하면 부족한 돈
## [대융금] = 보유하고 있는 주식을 일정한 비율로 환산한 금액
## ------------------------------------------------------------------------------------
## 11. 투자의 종류
### 1. 가치투자 : 회사의 가치를 보고 투자
### 2. 차트투자 : 차트의 흐름을 읽고 하는 투자
### 3. 심리투자 : 호재, 정보, 이슈 등으로 하는 투자
## ------------------------------------------------------------------------------------
## 12. 주식 전문 용어
## [PER] 주가수익률 : Price Earning Rstio
###  = 기업의 현재 시가총액 / 주당순이익
###  -> 기업의 고평가/저평가 여부를 알 수 있는 지표
###  ex) PER 높다 = 고평가, PER 낮다 = 저평가
### 주의! R&D, 연구, 개발, 벤처의 경우 PER이 주로 높음(잠재력)
###      우량주일 경우 대부분 PER이 낮음(기업의 규모가 크기 때문)
### 따라서, 해당 산업과 기업의 폭넓은 관점으로 이해 필요!
## [PBR] 주가준자산비율
### PBR < 1 : 저평가
### PBR > 1 : 고평가
### 따라서, PBR 10은 매우 고평가 된 종목
## [ROE] 자기자본이익률
### ROE = 당기순이익 / 자기자본 X 100
### 3년 연속 ROE가 15% 이상 되는 기업을 선호(워렌버핏)
## ------------------------------------------------------------------------------------
## 13. 주식은어
### ★ 개미털기 : 세력이 개미들이 접근하지 못하도록 시세 변동을 유도하는 행위
### ★ 외계인 : 외국인투자자, 외국계투자자
### ★ 흙두루미 : 세상 물정 모르는 사람
### ★ 하이에나 : 특정 종목을 집중매집하여 매무세가 몰리면 단기 급상승 했을 때 털고 나오는 세력
### ★ 깡통 : 빈 계좌
### ★ 우주방어 : 주가가 하락하지 않도록 기를 쓰고 막는 것
## ------------------------------------------------------------------------------------
## 14. 증자, 감자
### ★ 증자 : 주식을 발행해 회사의 자본금을 증가
### ★ 감자 : 주식 금액이나 주식 수의 감면 등을 통해 자본금을 감소
## ★ 유상증자 : 기업이 신규주식을 발행하여 기존 주주나 새로운 주주에게 팔아 자본을 증가시키는 수단
### 유산증자를 하는 이유 ? 
### - 회사운용에 필요한 자본을 조달
### - 투자를 위한 자금을 조달
## 자본조달의 대표적 3가지 방법
### 1. 은행대출
### 2. 채권발행
### 3. 유상증자(기업이 선호하는 자금 조달 방법. Why? 대출x, 이마 의무x, 자본금 상환x
## 유상증자의 장단점
### 장점 : 기업은 부담 없이 자본을 마련하기 좋다, 신규 투자자들은 괜찮은 가격에 매수 가능.
### 단점 : 시중의 주식을 대량으로 늘리면 주가는 떨어지게 된다. 기존 투자자들은 자신이 보유한 주식의 가격이 떨어짐.
### 결론 : 주가는 하락하지만, 다른 신규사업의 투자를 위한 유상증자일 경우, 호재 가능성 있다.
## ★ 무상증자 : 주식을 발행하여 주주들에게 무료로 나눠주는 것
### 자기자본 = 자본금 + 이익잉여금
### 잉여금을 자본금으로 옮기는 것. 사실상 자기자본금엔 변화는 없지만 이익 잉여금 ↓, 자본금 ↑
### 무상증자를 하는 이유?
### 주가를 관리하기 위해 무상증자를 함. 즉 인기 이미지 관리
### 결론 : 주주들에게 무상증자를 해주면, 주주들 입장에선 주식의 수가 늘어나고, 회사의 이미지 개선


```
