## ★죽을 때까지 매매해서는 안되는 종목들
### 1. 동전주 : 1,000원 미만의 주식을 뜻함. 상장폐지 되는 종목의 90%가 동전주.
### 2. 시가총액 200억 미만의 주식 : 시가총액이 작을 수록 회사가 망할 확률이 높음.
### 3. 우선주, 환기, 관리 : 우선주도 가급적 패스.('관','환' 제외)
### 4. 자본잠식 종목 : 기업분석창에 부채비율/유보율 란에 N/A라는 표시가 있는 종목은 매매 금지.
### 5. 회사의 존폐와 관련된 뉴스가 나오는 주식 : 1) 워크아웃 2) 횡령, 배임 3) 부도설
### ex) 대우조선해양 '워크아웃' 뉴스 출현(2015.07.15)

## 스윙매매(가치투자)
### - 장점 : 비교적 적은 횟수 판단으로 적은시간의 사용으로 높은 수익률을 얻을 수 있음
###       -> 5 ~ 10종목 많게는 10~ 30종목을 포트로 구성하여 길게 끌고 가야 한다. 
### - 단점 : 수익률은 단기매매보다 천천히 늘어 난다는 것도 인정해야 한다.

## 단기매매(차트투자)
### - 장점 : 시장의 리스크를 피해갈 수 있다.(즉 지수의 영향을 크게 받지 않는다.) 
### - 단점 : 시장에 꾸준히 참여해야 하기 때문에 시간은 반드시 써야 한다.
### 스윙과 단기의 기준 설정 모든 매매는 나의 성향과 환경을 담아야 한다.
## 추천하는 공부법
### - 단기적 성향, 여유로운 시간 
###   -> 수익률 10 ~ 20% 스윙(1주일 ~ 1달)
###   -> 수익률 35% 단기(1~5일)
### - 장기적 성향, 한정된 시간 
###   -> 수익률 30~50~100% 스윙(3개월 ~ 1년)
###   -> 수익률 10~ 20% 단기(1주일 ~ 1달)

## 모든 매매법에서 필요한 기준 5가지

### 1. 모든 매매에는 100%가 없다. 우리는 절대 신이 아님을 명심하자.
###   -> 지수차트 확인 후 반등이 될 것 같은 시점에 투자 지수는 23일 상승, 23일 하락(40년 확률)
### 2. 시장을 이기고 있는가? = 코스피 지수보다 강한 종목을 매매한다.(누군가 관리를 한다) 
###   -> ex) 코로나 이후 녹십자, 셀트리온, Naver, kakao
### 3. 주도주, 대장주인가? 우리의 돈은 한정적이다, 그리고 시장의 돈도 한정적이다. 
###   -> ex) 코로나 이후 녹십자 대장주 다음 Naver 주도주 : 당일날 더 많이 오르고, 덜 빠지는 주
