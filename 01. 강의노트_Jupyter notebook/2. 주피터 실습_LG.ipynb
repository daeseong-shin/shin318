### 0.2 2가지 모드
- Edit Mode (초록색): 셀 안의 내용을 편집할 수 있는 상태 : 셀 위에서 Enter 키를 눌러 셀 내용을 편집할 수 있음
- Command Mode (파란색): 셀 밖에서 해당 노트북을 명령어로 컨트롤할 수 있는 상태 : 셀 위에서 명령을 내릴 수 있음
- h를 눌러 도움말을 볼 수 있다.

### 0.3 유용한 커맨드 모드 명령어
- CTRL + Enter: 셀 실행
- Shift + Enter : 셀 실행후 다음 셀로 이동
- y: 셀을 Code 블록으로 변환
- m: 셀을 Markdown 블록으로 변환
- a: 위에 셀 추가
- b: 아래에 셀 추가
- dd: 셀 삭제
- 각자 위 명령어를 연습해보세요!!
